## Heston (scalability)

### Create

```python

Ns = [10, 32, 50, 100, 140]

# gp = lambda N: {
#     'name': 'NormalizerModel',
#     'kwargs': {
#         'model': {
#             'name': 'GPModel',
#             'kwargs': dict(
#                 kernel=dict(
#                     name='GPyRBF',
#                     kwargs={'lengthscale': 0.6, 'ARD': True},
#                 ),
#                 noise_prior=None,
#                 do_optimize=True,
#                 num_mcmc=0,
#             )
#         },
#     }
# }
gp = lambda N: {
    'name': 'NormalizerModel',
    'kwargs': {
        'model': {
            'name': 'DKLGPModel',
            'kwargs': {
                'learning_rate': 0.1,
                'n_iter': 150,
                'nn_kwargs': {'layers': None},
                'gp_kwargs': {'n_grid': None},
                'use_cg': False,
                'noise': None
            }
        }
    }
}

kissmodels = lambda N: {
    'name': 'NormalizerModel',
    'kwargs': {
        'model': {
            'name': 'DKLGPModel',
            'kwargs': {
                'learning_rate': 0.1,
                'n_iter': 150,
                'nn_kwargs': {'layers': None},
                'gp_kwargs': {'n_grid': N},
                'max_cg_iter': 1500,
                'precond_size': 10,
                'use_cg': True,
                'noise': None
            }
        }
    }
}

models = [gp, kissmodels]

for model in models:
    for N in Ns:
        run = execute(config_updates={
            'tag': 'heston',
            'obj_func': {
                'name': 'HestonOptionPricer',
            },
            'model': model(N),
            'gp_samples': N,
            'use_sample_grid': True,
        })

for d in range(1,5):
    run = execute(config_updates={
        'tag': 'heston',
        'obj_func': {
            'name': 'HestonOptionPricer',
        },
        'model': {
            'name': 'AdaptiveSparseGrid',
            'kwargs': dict(
                depth=d, 
                refinement_level=0,
                f_tol=1e-3,
            )
        },
    })
```

### View

```python
df = get_df()
exps_rows_df = df[df["tag"] == "heston"].copy()
exps_rows_df = exps_rows_df[exps_rows_df['id'] > 1000]
exps_rows_df
```
